# Slicing

- .loc performances label-based extraction.
- .iloc performs integer-based extraction.

In pandas, .loc and .iloc are two methods used to access and manipulate data in a DataFrame or Series by selecting specific rows and columns. They are powerful tools for selecting data based on labels (.loc) or index positions (.iloc).

.loc: Label-Based Indexing
.loc is used for selecting data by label (i.e., row and column names).
It allows you to select rows and columns based on their index labels (row/column names).
It can handle both single labels and lists or slices of labels.

.iloc: Position-Based Indexing
.iloc is used for selecting data by integer positions (i.e., row and column indices).
It allows you to access data based on the position of rows and columns rather than their labels.

```python
df.loc[0, 'Loan_ID']
```

```python
df.loc[[68, 100, 201], 'Married']
```

```python
df.iloc[0,4]
```

Group-wise Summary with df.groupby() and agg()
You can group data by one or more columns and summarize with aggregation functions like mean(), sum(), count(), etc.

```python
df.groupby('Education').agg({'ApplicantIncome': 'mean', 'LoanAmount': 'sum'})
```
