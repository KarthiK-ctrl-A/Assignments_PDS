```
import plotly.express as px

# Line plot for vaccines administered over time
fig1 = px.line(df, x='date', y='vaccines', title='Vaccines Administered Over Time in Australia',
               labels={'vaccines': 'Vaccines Administered', 'date': 'Date'}, template='plotly')

# Customize the layout for interactivity
fig1.update_layout(
    xaxis_title='Date',
    yaxis_title='Number of Vaccines Administered',
    hovermode='x unified',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=6, label="6m", step="month", stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(visible=True),
        type="date"
    )
)
fig1.show()
```

```
# Bar chart for tests conducted over time
fig2 = px.bar(df, x='date', y='tests', title='COVID-19 Tests Conducted Over Time in Australia',
              labels={'tests': 'Tests Conducted', 'date': 'Date'}, template='plotly')

# Customize the layout for interactivity
fig2.update_layout(
    xaxis_title='Date',
    yaxis_title='Number of Tests Conducted',
    hovermode='x unified',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=6, label="6m", step="month", stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(visible=True),
        type="date"
    )
)
fig2.show()
```
