```
import plotly.graph_objects as go

# Create the multi-line chart
fig1 = go.Figure()

# Add a line for cumulative confirmed cases
fig1.add_trace(go.Scatter(x=df['date'], y=df['confirmed_cum'], mode='lines', name='Cumulative Confirmed Cases'))

# Add a line for cumulative deaths
fig1.add_trace(go.Scatter(x=df['date'], y=df['deaths_cum'], mode='lines', name='Cumulative Deaths'))

# Add a line for cumulative vaccinations
fig1.add_trace(go.Scatter(x=df['date'], y=df['vaccines_cum'], mode='lines', name='Cumulative Vaccinations'))

# Customize the layout
fig1.update_layout(
    title='Cumulative COVID-19 Metrics Over Time (Confirmed Cases, Deaths, Vaccinations)',
    xaxis_title='Date',
    yaxis_title='Cumulative Count',
    hovermode='x unified',  # Hovering shows data across all lines
    legend_title='Metrics',
    template='plotly_dark',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=6, label="6m", step="month", stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(visible=True),
        type="date"
    )
)

fig1.show()
```

```
fig2 = go.Figure()

# Add an area trace for cumulative hospitalizations
fig2.add_trace(go.Scatter(x=df['date'], y=df['hosp_cum'], fill='tozeroy', mode='none', name='Cumulative Hospitalizations'))

# Add an area trace for cumulative ICU admissions
fig2.add_trace(go.Scatter(x=df['date'], y=df['icu_cum'], fill='tonexty', mode='none', name='Cumulative ICU Admissions'))

# Add an area trace for cumulative ventilator usage
fig2.add_trace(go.Scatter(x=df['date'], y=df['vent_cum'], fill='tonexty', mode='none', name='Cumulative Ventilator Usage'))

# Customize layout
fig2.update_layout(
    title='Cumulative Hospitalizations, ICU Admissions, and Ventilator Usage Over Time',
    xaxis_title='Date',
    yaxis_title='Cumulative Count',
    hovermode='x unified',
    template='plotly_dark',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=6, label="6m", step="month", stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(visible=True),
        type="date"
    )
)

fig2.show()
```
